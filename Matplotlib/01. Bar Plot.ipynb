```python
"""
Bar plot adalah plot yang digunakan untuk menampilkan data kategorikal.
"""
import matplotlib.pyplot as plt
import numpy as np
```

## vertikal

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 20, 30, 40])
z = np.array([5, 15, 25, 35])
p = np.arange(len(x))
lebar = 0.3
p1 = [j+lebar for j in p]
plt.bar(p, y,width=lebar,color = "y",align='center',label = "penjualan 1") # bar plot berdasarkan data x dan y
plt.bar(p1, z,width=lebar,color = "g",align='center',label = "penjualan 2") # bar plot berdasarkan data x dan z
plt.xlabel("Kategori") # label sumbu x
plt.ylabel("Jumlah") # label sumbu y
plt.xticks(p +lebar/2, x,rotation = 10) # label sumbu x
plt.title("Contoh Bar Plot") # judul plot
plt.legend()
plt.show() # menampilkan plot
```

## horiontal

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 20, 30, 40])
z = np.array([5, 15, 25, 35])
p = np.arange(len(x))
lebar = 0.3
p1 = [j+lebar for j in p]
plt.barh(p, y, lebar,color="y", align='center', label="penjualan 1") # bar plot berdasarkan data x dan y
plt.barh(p1, z, lebar,color="g", align='center', label="penjualan 2") # bar plot berdasarkan data x dan z
plt.xlabel("Kategori") # label sumbu x
plt.ylabel("Jumlah") # label sumbu y
plt.title("Contoh Bar Plot") # judul plot
plt.yticks(p + lebar/2, x, rotation=10) # label sumbu x
plt.legend()
plt.show() # menampilkan plot
```
