```python
"""
Bar plot adalah plot yang digunakan untuk menampilkan data kategorikal.
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 20, 30, 40])
plt.bar(x, y,width=0.9,color = "y") # bar plot berdasarkan data x dan y
plt.xlabel("Kategori") # label sumbu x
plt.ylabel("Jumlah") # label sumbu y
plt.title("Contoh Bar Plot") # judul plot
plt.show() # menampilkan plot
```
