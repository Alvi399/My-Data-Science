```python
"""
Histogram plot adalah plot yang menampilkan distribusi frekuensi dari data numerik
dalam bentuk batang-batang vertikal atau horizontal. Histogram plot biasanya
digunakan untuk menampilkan distribusi frekuensi dari data numerik yang terbagi
menjadi beberapa kelas atau interval.
"""
import matplotlib.pyplot as plt
import numpy as np
```

## vertikal

```python
x = np.random.randint(10,60,50)

plt.hist(x,color='green',alpha=0.5,edgecolor='black',bins=10,align="left",histtype="barstacked",rwidth=0.8) #bins adalah jumlah batang, align adalah posisi batang, hsittype adalah tipe batang
plt.title('Contoh Histogram Plot',fontsize=15)
plt.xlabel('Data',fontsize=12)
plt.ylabel('Frekuensi',fontsize=12)
plt.axvline(x.mean(),color='red',label="rata-rata",linestyle="--",linewidth=2) #menambahkan garis vertikal
plt.legend()
plt.grid(True)
plt.show()
```

## horizontal


```python
x = np.random.randint(10,60,50)

plt.hist(x,color='green',alpha=0.5,edgecolor='black',bins=10,align="left",histtype="barstacked", orientation="horizontal",rwidth = 0.8,log=True,label="data penduduk") #bins adalah jumlah batang, align adalah posisi batang, hsittype adalah tipe batang
plt.title('Contoh Histogram Plot',fontsize=15)
plt.xlabel('Data',fontsize=12)
plt.ylabel('Frekuensi',fontsize=12)
plt.grid()
plt.legend()
plt.show()
```
