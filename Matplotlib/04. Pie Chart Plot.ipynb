```python
"""
Pie Chart adalah grafik lingkaran yang menunjukkan proporsi dari data yang berbeda.
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([35, 25, 25, 15])
y = np.array(["Apple", "Banana", "Cherry", "Dates"])
e = np.array([0.1, 0, 0, 0])
color = np.array(["red", "green", "blue", "yellow"])

plt.pie(
    x=x, 
    labels=y,
    explode=e,colors=color,
    autopct='%1.1f%%',
    shadow=True,
    radius=1,
    labeldistance= 1.1,
    startangle=30, 
    textprops={'fontsize': 10},
    counterclock=True,
    wedgeprops=dict( linewidth=2, edgecolor="w"),
    rotatelabels=False
        ) # autopct='%1.1f%%' untuk menampilkan persentase, labeldistance= 1.1 untuk jarak label dari lingkaran, startangle=90 untuk memulai dari sudut 90 derajat
#textprops={'fontsize': 10} untuk mengatur ukuran font
#counterclock=False untuk mengatur arah jarum jam
#wedgeprops=dict( linewidth=2, edgecolor="w") untuk mengatur garis tepi
plt.title("Contoh Pie chart")
plt.legend(title="Fruit", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1)) #parameter bbox_to_anchor untuk mengatur posisi legend
plt.show()
```
