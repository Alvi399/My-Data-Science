```python
"""
Steam plot adalah plot yang digunakan untuk menampilkan data yang berbentuk garis lurus
bisanya digunakan untuk menampilkan data yang berbentuk time series data
"""
import matplotlib.pyplot as plt
import numpy as np
```

## Vertikal

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 25, 30, 40])

plt.stem(
    x, y, 
    linefmt = 'b:', 
    markerfmt = 'ro',
    bottom=0,
    basefmt = 'g-',
    label="jangkauan waktu",
)
#linefmt adalah format garis
#markerfmt adalah format marker
#basefmt adalah format garis dasar
plt.ylabel('Data',fontsize = 15)
plt.xlabel('Waktu (s)',fontsize = 15)
plt.title('Contoh Steam Plot',fontsize = 15)
plt.legend()
plt.show()
```

## Horizontal

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 25, 30, 40])

plt.stem(
    x, y, 
    linefmt = 'b:', 
    markerfmt = 'ro',
    bottom=0,
    basefmt = 'g-',
    label="jangkauan waktu",
    orientation='horizontal'
)
#linefmt adalah format garis
#markerfmt adalah format marker
#basefmt adalah format garis dasar
plt.xlabel('Data',fontsize = 15)
plt.ylabel('Waktu (s)',fontsize = 15)
plt.title('Contoh Steam Plot',fontsize = 15)
plt.legend()
plt.show()
```
