```python
"""
Stack Plot adalah plot yang menunjukkan bagaimana setiap nilai dalam kumpulan data berkontribusi terhadap keseluruhan nilai.
biasanya digunakan dalam data yang berulang seperti data waktu. 
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
area = np.array([2, 5, 6, 4, 3])
area2 = np.array([1, 2, 1, 1, 1])
area3 = np.array([2, 3, 3, 3, 3])

plt.stackplot(
    x, area,area2,area3,
    labels=['A', 'B', 'C', 'D', 'E'],
    colors=['r', 'g', 'b', 'y', 'm']
)
plt.legend(loc=2) #parameter loc = 2 berarti posisi legend berada di kiri atas
plt.grid(True)
plt.show()
```
