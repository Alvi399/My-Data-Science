```python
"""
Step Plot adalah plot yang menampilkan data dalam bentuk step. Step plot digunakan untuk menampilkan data yang berubah secara diskrit, 
seperti data yang diambil dari sensor atau data yang diambil dari eksperimen.
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 5, 6, 7, 8])
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([3, 4, 5, 6, 7])

plt.step(
    x=x,
    y=y,
    marker='o',
    ms = 10,
    mfc = 'r',
    label='Data'
)
plt.step(
    x=x2,
    y=y2,
    marker='x',
    ms = 10,
    mfc = 'g',
    label='Data 2'
)
#parammeter ms adalah ukuran marker
#parameter mfc adalah warna marker
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Contoh Step Plot')
plt.grid(True)
plt.show()
```
