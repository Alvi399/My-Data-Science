```python
"""
Fiil between plot adalah untuk mengisi area antara dua garis atau kurva pada plot dengan warna atau pola tertentu.
biasanya digunakan untuk menunjukkan area tertentu pada plot yang memiliki makna tertentu.
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([4, 3, 5, 6, 3])

plt.plot(x, y1, marker='o')
plt.fill_between(
    x=x,
    y1=y1,
    y2=0,
    color='orange',
    where=(x >= 2) & (x <= 4),
    alpha=0.3
)
plt.title("Contoh Fill Between Plot")
plt.xlabel("Sumbu X",fontsize =15)
plt.ylabel("Sumbu Y", fontsize =15)
plt.show()
```
