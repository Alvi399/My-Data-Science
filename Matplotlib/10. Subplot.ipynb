```python
"""
Subplot adalah cara untuk menampilkan lebih dari satu plot dalam satu figure.
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Membuat plot
plt.subplot(2, 2, 1)
plt.plot(x, y)

#membuat pie
plt.subplot(2, 2, 2)
plt.pie(y)

#membuat scatter
plt.subplot(2, 2, 3)
plt.scatter(x, y)


#mebuat bar
plt.subplot(2, 2, 4)
plt.bar(x, y)
plt.show()

#format subplot
#plt.subplot(1, 2, 1) # 1 baris, 2 kolom, posisi 1
```
