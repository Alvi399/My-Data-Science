```python
import matplotlib.pyplot as plt
import numpy as np
```

## Axes

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])

plt.plot(x, y)
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['Python', 'Java', 'C++', 'C#', 'JavaScript'])
plt.yticks(ticks=[2, 3, 4, 5, 6], labels=['2', '3', '4', '5', '6'])
plt.xlim(1,6)
plt.ylim(3,6)
plt.axis([1, 6, 3, 8])
#xticks digunakan untuk mengatur label pada sumbu x
#yticks digunakan untuk mengatur label pada sumbu y
#xlim digunakan untuk mengatur batas sumbu x
#ylim digunakan untuk mengatur batas sumbu y
#axis digunakan untuk mengatur batas sumbu x dan y
plt.show()
```

## Text

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])
plt.plot(x, y)
plt.title('Grafik Linear')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.text(x=3, y=4, s='Titik (3,4)',style='italic',bbox=dict(facecolor='red', alpha=0.5))
plt.annotate('Titik Penting', xy=(3, 4), xytext=(3.5, 4.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(["Legend"],loc = 1,facecolor = "yellow",edgecolor = 'red',shadow = True)
plt.show()
```
