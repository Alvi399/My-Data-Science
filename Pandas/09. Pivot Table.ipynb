```python
import pandas as pd
data = pd.read_csv("Example Dataset/weather.csv")
data
```

## Using Pivot()

```python
#pivot adalah mengubah bentuk data
#mengubah bentuk data dari baris menjadi kolom
#mengubah bentuk data dari kolom menjadi baris

new_data = data.pivot(index="date", columns="city")
new_data
```

```python
new_data = data.pivot(index ="temperature", columns="city")
new_data
```

## Using pivot_table()

```python
#pivot_table adalah mengubah bentuk data
#prbedaan pivot dan pivot_table adalah pivot_table bisa mengubah data yang sama
new_data = data.pivot_table(index="city", columns="date",margins=True) #parameter margins=True untuk menambahkan kolom dan baris total
new_data
```

```python
new_data = data[data.city == "mumbai"]
new_data['date'] = pd.to_datetime(new_data["date"])
new_data = new_data.pivot_table(index=pd.Grouper(freq="M", key="date"), columns="city") #parameter freq="M" untuk mengubah data menjadi bulanan
new_data
```
