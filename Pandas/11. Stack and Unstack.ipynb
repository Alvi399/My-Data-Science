```python
import pandas as pd
df = pd.read_excel("Example Dataset/stocks.xlsx",header=[0,1]) #parameter header berfungsi untuk menggabungkan 2 baris header
df
```

## Using stack()

```python
data = df.stack() #mengubah kolom menjadi baris
data
```

## Using unstack()

```python
data = df.unstack() #mengubah baris menjadi kolom
data
```

## Using Stack 3 Level

```python
df = pd.read_excel("Example Dataset/stocks_3_levels.xlsx",header=[0,1,2]) #parameter header berfungsi untuk menggabungkan 3 baris header
df
```

```python
df = df.stack() #mengubah kolom menjadi baris
df
```

## Using Unstack 3 Level

```python
df = df.unstack() #mengubah baris menjadi kolom
df
```
