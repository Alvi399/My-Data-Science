```python
import pandas as pd
import numpy as np
data = pd.read_excel("Example Dataset/survey.xls")
data.tail()
```

## Using crosstab()

```python
#crosstab adalah sebuah fungsi yang dapat digunakan untuk membuat tabel silang (cross tabulation) pada Pandas.
#cross tabulation adalah sebuah tabel yang dapat digunakan untuk menghitung frekuensi dari satu atau lebih variabel.
df = pd.crosstab(data["Nationality"], data["Handedness"])
df
```

```python
df = pd.crosstab(data["Sex"], data["Handedness"],normalize="index") #normalize="index" untuk mengubah nilai menjadi persentase
df
```

```python
df = pd.crosstab(data["Sex"], data["Handedness"],margins=True) #margins=True untuk menambahkan subtotal
df
```

```python
a = 'sum'
df = pd.crosstab([data["Nationality"],data["Age"]], data["Handedness"], values=data["Handedness"], aggfunc=a, margins=True) #values=data["Handedness"] untuk mengisi nilai tabel dengan kolom Handedness
df
```

```python
df = pd.crosstab(data["Sex"],data["Handedness"], values=data["Age"], aggfunc=np.average, margins=True) #values=data["Age"] untuk mengisi nilai tabel dengan kolom Age                        
df
```
