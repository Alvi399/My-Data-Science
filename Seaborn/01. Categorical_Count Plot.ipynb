```python
"""
Count Plot adalah plot yang digunakan untuk menghitung jumlah data yang ada di dalam suatu kolom.
kalau dalam matplotlib kita menggunakan method plt.hist() untuk membuat histogram, 
maka dalam seaborn kita menggunakan method sns.countplot() untuk membuat count plot.
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('src/supermarket_sales.xlsx')
df.head()
```

## 1. Basic Countplot

```python
plt.figure('Count Plot', figsize=(16, 5))
sns.countplot(x='Product line', data=df, palette='viridis')
```

## 2. Make it Horizontal

```python
sns.countplot(y='Product line', data=df, palette='viridis')
```

## 3. Add hue to get count on two categoris

```python
plt.figure('Count Plot', figsize=(16, 5))
sns.countplot(x = 'Product line',hue='Gender', data=df, palette='viridis')
# parameter hue digunakan untuk membagi data berdasarkan kategori tertentu
```

## 4. Using color pallate

```python
plt.figure('Count Plot', figsize=(16, 5))
sns.countplot(x = 'Product line',hue='Payment', data=df, palette='viridis')
```

```python
plt.figure('Count Plot', figsize=(16, 5))
sns.countplot(x = 'Product line',hue='City', data=df, palette='gist_rainbow')
```

## 5. Styling bar with forecolor, edgecolor, and linewidth

```python
plt.figure('Count Plot', figsize=(16, 5))
sns.countplot(
    x = 'Product line',
    data=df,
    palette='gist_rainbow',
    facecolor= (1, 0, 1, 0),
    linewidth=5,
    edgecolor=sns.color_palette('dark', 3)
)
#parameter facecolor digunakan untuk mengatur warna dari area dalam plot
#parameter linewidth digunakan untuk mengatur ketebalan garis batas
#parameter edgecolor digunakan untuk mengatur warna dari garis batas
```
