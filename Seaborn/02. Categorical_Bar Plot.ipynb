## 2. Basic Barplot

```python
df = pd.read_excel('src/supermarket_sales.xlsx')
df.head(3)
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total",hue= "Gender", data=df,palette='viridis')
plt.show()
#garis hitam adalah confidence interval atau batas kepercayaan dari data yang diambil
#garis hitam tersebut menunjukkan bahwa data yang diambil adalah data yang valid atau bisa dipercaya
#dari data diatas kita bisa melihat bahwa produk yang paling banyak terjual adalah produk fashion dan produk yang paling sedikit terjual adalah produk electronic
#dan dari data diatas kita juga bisa melihat bahwa pria lebih banyak membeli produk fashion dan produk health and beauty sedangkan wanita lebih banyak membeli produk electronic dan produk home and lifestyle
```

## Horizontal Barplot

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Total", y="Product line",hue= "Payment", data=df,palette='viridis',)
```

## Given Order in plotbar

```python
x = df["Product line"].sort_values()
x.unique()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total",hue= "Payment", data=df,palette='viridis',order=x,hue_order=["Ewallet","Cash","Credit card"])
plt.show()
```

## add CAP in eror bar

```python
#eror bar adalah garis yang menunjukkan batas kepercayaan dari data yang diambil
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total",hue= "Branch", data=df,palette='viridis',order=x,hue_order=["A","B","C"],capsize=0.1)
#parameter cipsize digunkan untuk menambahkan garis error
plt.show()
```

## Remove erorbar

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total",hue= "Customer type", data=df,palette='viridis',order=x,hue_order=["Member","Normal"],ci=0)
#parameter ci digunakan untuk menghilangkan garis error
plt.show()
```

## Change color and saturation

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total", data=df,color='red',saturation=0.001)
plt.show()
```

## Change Default Aggregation Method

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="Product line", y="Total", data=df,color='red',hue="Gender",saturation=0.001,estimator=np.median) 
#estimator digunakan untuk menghitung jumlah total penjualan
plt.show()
```
