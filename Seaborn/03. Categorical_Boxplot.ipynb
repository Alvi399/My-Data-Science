```python
"""
Boxplot adalah cara yang bagus untuk membandingkan beberapa kelompok dan melihat distribusi mereka.
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_excel('src/supermarket_sales.xlsx')
df.head(3)
```

## Basic Boxplot


```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df, x='Product line', y='Total')
```

## Basic Boxplot On Numeric Variabel

```python
sns.set_theme(style="darkgrid")
sns.boxplot(y = 'Total',data=df,width=0.3)
```

## Basic Boxplot on Categorical Variabel

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x = 'Product line',y = 'Total',data=df)
```

## Basic Boxplot in two Categorical Variabel

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x = 'Product line',y = 'Total',data=df, hue='Gender')
```

## add Mean Marker

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x = 'Product line',y = 'Total',data=df,showmeans=True,meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black"})
```

## Horizontal Boxplot

```python
sns.boxplot(x = "Total", y = "Product line", data = df, orient = "h",hue="Gender")
```
