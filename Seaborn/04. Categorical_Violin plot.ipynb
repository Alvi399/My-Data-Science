```python
"""
Violin Plot adalah gabungan dari box plot dan density plot. 
Violin plot menggambarkan distribusi data dan juga menggambarkan distribusi data pada setiap level kategori. 
Violin plot biasanya digunakan untuk membandingkan distribusi data pada beberapa level kategori.
Violin plot dapat digambarkan dengan menggunakan fungsi `violinplot()` dari library seaborn. 
Berikut adalah contoh penggunaan violin plot untuk membandingkan distribusi data pada beberapa level kategori.
"""
import seaborn as sns
import pandas as pd
df = pd.read_excel("./src/supermarket_sales.xlsx")
df.head(3)
```

## Basic Violin Plot

```python
sns.violinplot(y="Total",data=df)
```

## Violin Plot two categorical

```python
sns.violinplot(data=df,x="Payment",y="Total",palette="muted",hue="Gender")
```

## Change Violinlot to horizontalline

```python
sns.violinplot(x="Payment",y="Total",hue="Gender",data=df,palette="muted",inner="quartile",split=True)
#paamerter split=True digunakan untuk memisahkan distribusi data pada setiap level kategori
```

```python
sns.violinplot(x="Payment",y="Total",hue="Gender",data=df,palette="muted",inner="stick",split=True)
```
