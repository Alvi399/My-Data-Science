```python
"""
Stripplot adalah plot yang digunakan untuk menampilkan data dalam bentuk titik-titik.
fungsi strip plot adalah untuk menampilkan data dalam bentuk titik-titik yang menggambarkan distribusi data dari setiap kategori.
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_excel("./src/supermarket_sales.xlsx")
df.head(3)
```

## Basic Strip Plot

```python
sns.stripplot(data=df, y='Total')
```

```python
sns.stripplot(
    data=df,
    y='Total',
    jitter=0.3,
    linewidth=1
    ) 
# jitter adalah jarak antar titik
# linewidth adalah ketebalan garis
```

## Using More Two Variabel

```python
sns.stripplot(
    data=df,
    x='Payment',
    y='Total',
    jitter=0.3,
    linewidth=1,
)
```

```python
sns.stripplot(
    data=df,
    x='Payment',
    y='Total',
    hue='Gender',
    jitter=0.3,
    linewidth=1,
)
```

```python
sns.stripplot(
    data=df,
    x='Payment',
    y='Total',
    hue='Gender',
    jitter=0.3,
    linewidth=1,
    dodge=True
)
#parameter doge adalah untuk memisahkan data berdasarkan kategori
```

# Draw Strip Plot on OtherPlot


```python
sns.stripplot(
    data=df,
    x='Payment',
    y='Total',
    jitter=0.3,
    linewidth=1,
)
sns.violinplot(
    data=df,
    x='Payment',
    y='Total',
    inner=None
)
```

```python
sns.stripplot(
    data=df,
    x='Payment',
    y='Total',
    jitter=0.3,
    linewidth=1,
)
sns.boxplot(
    data=df,
    x='Payment',
    y='Total'
)
```
