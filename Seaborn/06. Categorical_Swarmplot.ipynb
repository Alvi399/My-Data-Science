```python
"""
Swrmplot adalah plot untuk menampilkan data dalam bentuk swarmplot
bisanya digunakan untuk menampilkan data kategorikal
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_excel('./src/supermarket_sales.xlsx')
df.head(3)
```

## Basic Swarmplot

```python
sns.swarmplot(data=df,y="Total")
```

```python
sns.swarmplot(data=df,x="Payment",y="Total",palette="Set2")
```

```python
sns.swarmplot(data=df,x="Payment",y="Total",hue="Gender",palette="Set2",dodge=True)
```

## Advanced Swarmplot

```python
sns.swarmplot(x="Payment",y="Total",hue="Gender",data=df,palette="Set2",dodge=True,marker="*",size=10)
```

```python
sns.boxplot(x="Payment",y="Total",data=df,palette="Set2")
sns.swarmplot(x="Payment",y="Total",data=df,color="black",size=3)
```

```python
sns.violinplot(x="Payment",y="Total",data=df,palette="Set2")
sns.swarmplot(x="Payment",y="Total",data=df,color="black",size=3)
```
