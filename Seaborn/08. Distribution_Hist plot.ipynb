```python
"""
Histrogram adalah sebah grafik yang menampilkan distribusi frekuensi dari suatu data.
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_excel('./src/supermarket_sales.xlsx')
df.head(3)
```

## Basic Hist Plot

```python
sns.histplot(data=df,x='gross income',kde=False)
```

```python
sns.histplot(data=df,y='gross income',kde=True)
```

## Advanced Hist Plot

```python
# Change number of bins and interval
sns.histplot(data=df,x='gross income',bins=20)
```

```python
sns.histplot(data=df,x='gross income',bins=np.arange(0,100,5))
plt.xticks(np.arange(0,100,5))
```

```python
# combine with hue
sns.histplot(data=df,x='Total', bins=20, hue = 'Payment' )
```

```python
# combine with multiple hue
sns.histplot(data=df,x='Total', bins=20, hue = 'Payment', multiple='stack' )
#fungsi parameter multiple adalah untuk menggabungkan hue menjadi satu grafik
#multiple='stack' berarti menggabungkan hue menjadi satu grafik yang ditumpuk
#multiple='dodge' berarti menggabungkan hue menjadi satu grafik yang berdampingan
#jika tidak diisi maka hue akan menjadi satu grafik yang berdampingan
```

```python
#make poly in element
sns.histplot(data=df,x='Total', bins=20, element='poly',hue='Payment')
```

```python
#create a bivarete histogram
sns.histplot(data=df,x='Total', y='gross income', bins=20, cbar=True)
#parameter cbar=True adalah untuk membuat color bar pada grafik
```
