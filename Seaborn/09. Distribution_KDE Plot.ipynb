```python
"""
KDE plot adalah
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_excel("./src/supermarket_sales.xlsx")
df.head(3)
```

## Basic KDE Plot

```python
sns.kdeplot(data=df)
```

```python
sns.kdeplot(data=df, x="Total", shade=True,bw_adjust=0.5)
#parameter bw_adjust untuk mengatur lebar garis
```

## Adjust smoothing using bw_adjust 

```python
sns.kdeplot(data=df, x="Total",bw_adjust=0.1)
```

## Group the KDE on a category variabel

```python
sns.kdeplot(data=df, x="Total",bw_adjust=0.5, fill=True, hue="Payment",multiple='stack')
```

## Use log scalling to map the variabelin KDE

```python
sns.kdeplot(data=df,x="Total",log_scale=True)
```

## Cghange styling od head KDE using linewidth, pallate, etc

```python
sns.kdeplot(data=df,x="Total",hue="Payment",multiple="stack",linewidth=1, palette="tab10")
```

## Creat bivarete KDE

```python
sns.kdeplot(data=df,x="Unit price",y="Total")
```

## Create bivarete with on a categorcal variabel

```python
sns.kdeplot(data=df,x="Unit price",y="Total",fill=True,thresh=0.2,levels=5,cmap="mako")
#parameter thresh untuk mengatur batas level
#lavel untuk mengatur jumlah level
# cmap untuk mengatur warna
```
