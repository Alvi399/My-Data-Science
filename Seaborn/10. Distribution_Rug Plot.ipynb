```python
"""
Rugplot adalah ebuah plot yang menunjukkan data dalam bentuk garis horizontal. 
Rugplot adalah plot yang sederhana dan berguna untuk menunjukkan distribusi data dalam satu dimensi. Rugplot adalah plot yang sederhana dan berguna untuk menunjukkan distribusi data dalam satu dimensi. 
Rugplot adalah plot yang sederhana
"""
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
df = pd.read_excel("./src/supermarket_sales.xlsx")
df.head(3)
```

## Create Basic RugPlot 

```python
#one variabel
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income')
```

```python
#two variabel
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income', y='Unit price')
```

## Group by Categorical variabel using hue

```python
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income', y='Unit price', hue='Gender')
```

## Change the height of right in the RUG Plot

```python
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income', y='Unit price', hue = 'Gender',height=0.5)
```

## Combine with KDE Plot

```python
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income', y='Unit price', hue = 'Gender')
sns.kdeplot(data=df, x = 'gross income', y = 'Unit price', hue= 'Gender')
```

## Comine with scatter plot

```python
plt.figure(figsize=(10,5))
sns.rugplot(data=df, x='gross income', y='Unit price', hue = "Gender")
sns.scatterplot(data=df, x='gross income', y='Unit price', hue= 'Gender')
```

## Show out area

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='gross income', y='Unit price', hue = 'Gender')
sns.rugplot(data=df, x='gross income', y='Unit price', hue = 'Gender',clip_on=False, height=-0.05)
#clip_on=False, maka data yang berada di luar plot akan tetap ditampilkan
```
