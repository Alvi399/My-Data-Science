```python
"""
Displot adalah seuah plot yang digunakan untuk menampilkan distribusi data
"""
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel("./src/supermarket_sales.xlsx")
df = df["Gender", "Product line", "Total", "Quantity", "Payment", "cogs", "gross income"]
df.head(3)
```

```python
sns.displot(data=df, x="gross income",kde = True,rug=True,rug_kws={"color":"r","height":0.05})
```

```python
sns.displot(
    data=df, 
    x="gross income",
    kde = True,
    rug=True,
    rug_kws={"color":"r","height":0.05},
    hue="Payment"
)
```

```python
sns.displot(
    data=df, 
    y="gross income",
    kde = True,
    rug=True,
    rug_kws={"color":"r","height":0.05},
    hue="Payment",
    col="Gender",
    row="Branch",
    multiple="stack"
)
```

```python
sns.displot(
    data=df, 
    y="gross income",
    kde = True,
    rug=True,
    rug_kws={"color":"r","height":0.05},
    hue="Payment",
    col="Gender",
    row="Branch",
    multiple="stack",
    kind="hist"
)
```
