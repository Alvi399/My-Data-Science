## Menyaipkan semua library yang dibuthkan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
data1_df = pd.read_csv('data/PRSA_Data_Changping_20130301-20170228.csv')
data2_df = pd.read_csv('data/PRSA_Data_Aotizhongxin_20130301-20170228.csv')
data3_df = pd.read_csv('data/PRSA_Data_Dingling_20130301-20170228.csv')
data4_df = pd.read_csv('data/PRSA_Data_Dongsi_20130301-20170228.csv')
data5_df = pd.read_csv('data/PRSA_Data_Guanyuan_20130301-20170228.csv')
data6_df = pd.read_csv('data/PRSA_Data_Gucheng_20130301-20170228.csv')
data7_df = pd.read_csv('data/PRSA_Data_Huairou_20130301-20170228.csv')
data8_df = pd.read_csv('data/PRSA_Data_Nongzhanguan_20130301-20170228.csv')
data9_df = pd.read_csv('data/PRSA_Data_Shunyi_20130301-20170228.csv')
data10_df = pd.read_csv('data/PRSA_Data_Tiantan_20130301-20170228.csv')
data11_df = pd.read_csv('data/PRSA_Data_Wanliu_20130301-20170228.csv')
data12_df = pd.read_csv('data/PRSA_Data_Wanshouxigong_20130301-20170228.csv')
```

```python
all_df = pd.concat(
    [data1_df,
    data2_df,
    data3_df,
    data4_df,
    data5_df,
    data6_df,
    data7_df,
    data8_df,
    data9_df,
    data10_df,
    data11_df,
    data12_df],
    ignore_index=False
).reset_index()
all_df.head()    
```

### Assessing Data

```python
all_df.drop
all_df['date'] = pd.to_datetime(all_df[['year', 'month', 'day']])
all_df = all_df = all_df.drop(['day', 'month', 'year'], axis=1)
all_df.head(12)
```

```python
all_df.info()
```

```python
all_df.isna().sum()
```

```python
print("Jumlah duplikasi: ",all_df.duplicated().sum())
```

```python
all_df.describe(include="all")
```

### Cleaning Data

```python
# Misssing value
columns =["PM2.5","PM10","SO2","NO2","CO","O3","TEMP","PRES","DEWP","RAIN","WSPM"]
for c in columns:
    all_df[c].fillna(value=all_df[c].mean(),inplace= True)
all_df["wd"].fillna(value="unknown",inplace=True)
all_df.isna().sum()
```

```python
all_df.drop(columns="index",axis=1,inplace=True)
all_df.describe(include="all")
```

## Exploratory Data Analysis (EDA)

### Explore All_df

```python
#Suhu rata - rata dalam lima tahun
temp_mean_df = all_df.groupby(by=["date"]).agg({
    "TEMP": "mean"
})
temp_mean_df.head(15)
```

```python
pm25_df = all_df.groupby(by="station").agg({
    "PM2.5":  "mean"
}).sort_values(by="PM2.5",ascending=False).reset_index()
pm25_df.head()
```

```python
pm10_df = all_df.groupby(by="station").agg({
    "PM10":  "mean"
}).sort_values(by="PM10",ascending=False).reset_index()
pm10_df.head()
```

## Visualization & Explanatory Analysis

### Rata tempertur suhu bulunan wilayah 

```python
temp_mean_monthly_df = all_df.resample(rule='M',on='date').agg({
    "TEMP":"mean"
}).reset_index()
temp_mean_monthly_df.head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(temp_mean_monthly_df['date'],temp_mean_monthly_df['TEMP'], marker= 'o',linewidth = 2,color = "#72BCD4")
plt.title("Diagram temperatur bulanan semua wilayh")
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15) 
plt.show()
```

### Tingkat polusi di beberapa wilyah

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize = (24,6))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="PM2.5", y="station", data=pm25_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Wilayah Polusi PM2.5 Terbanyak", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
sns.barplot(x="PM2.5", y="station", data = pm25_df.sort_values(by="PM2.5", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Wilayah Polusi PM2.5 Tersedikit", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
plt.suptitle("Wilayah Polusi PM2.5 Terbanyak dan Tersedikit", fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize = (24,6))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="PM10", y="station", data=pm10_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Wilayah Polusi PM10 Terbanyak", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
sns.barplot(x="PM10", y="station", data = pm10_df.sort_values(by="PM10", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Wilayah Polusi PM10 Tersedikit", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
plt.suptitle("Wilayah Polusi PM10 Terbanyak dan Tersedikit", fontsize=20)
plt.show()
```
