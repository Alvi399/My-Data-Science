```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Univariate visulation

## big Chart

```python

cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
    'Surakarta','Surabaya', 'Medan', 'Makassar')

populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
    3481, 287750, 785409)

plt.bar(x=cities, height=populations)
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.barh(y = cities,width= populations)
```

```python
df = pd.DataFrame({
    "cities": cities,
    "poulations": populations
})
df.head()
```

```python
sns.barplot(y = df["cities"],x=df["poulations"],orient="h",color="blue")
plt.xlabel("Population (Millions)")
plt.title("Population cities in indonesia")
plt.show()
```

## Pie chart

```python
flavor = ('apel','mangga','stroberi','sissak')
votes = (20,50,30,10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.1, 0.2, 0, 0)

plt.pie(
    x=votes,
    labels= flavor,
    colors=colors,
    wedgeprops={'width': 0.4},
    explode=explode,
    autopct='%1.1f%%'
)
plt.show()
```

## histogram

```python
x = np.random.normal(15,5,150)
plt.hist(x = x,bins=15)
plt.show()
```

# Bivariate and Multivariate Visulation

## Scatter Plot 

```python
import matplotlib.pyplot as plt
lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15, 
    9.96, 7.21, 10.04, 10.2, 11.06]
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93, 
    132.93, 138.92, 145.98, 148.44, 152.81]
plt.scatter(x=lemon_diameter, y=lemon_weight)
plt.show()
```

```python
sns.scatterplot(x = lemon_diameter,y=lemon_weight)
plt.show()
```

```python
sns.regplot(x = lemon_diameter,y=lemon_weight)
plt.show()
```

## Line Chart 

```python
plt.plot(lemon_diameter,lemon_weight)
plt.show()
```

```python
url =  'https://query1.finance.yahoo.com/v7/finance/download/BBCA.JK?period1=1644796800&period2=1676332800&interval=1d&events=history&includeAdjustedClose=true'
df = pd.read_csv(url)
df["Date"]= pd.to_datetime(df["Date"])
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Date'],df['Close'],colors='blue')
plt.xlabel('Date',size = 15)
plt.ylabel('Price',size =15)
plt.show()
```

## Clustered Bar Chart

```python

data = sns.load_dataset("penguins")
sns.barplot(data=data, x="species", y="body_mass_g", hue="sex", ci=None)
plt.show()
```

```python
df_boxpot = df[["Open","High","Low","Close","Adj Close"]]
sns.boxplot(data = df_boxpot, palette="rocket")
plt.ylabel("Price",size = 15)
plt.show()
```

```python
sns.scatterplot(data=data, x="body_mass_g", y="flipper_length_mm", hue="species", style="species")
plt.show()
```
